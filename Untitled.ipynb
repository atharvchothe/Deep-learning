```python
import numpy as np
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
np.random.seed(1)

# Input data (1 sample, 20 features)
x = np.random.randn(1, 20)

# Initialize weights and biases
weights = []
biases = []

for i in range(10):
    weights.append(np.random.randn(20, 20))
    biases.append(np.zeros((1, 20)))

# Forward pass
layer_output = x

for i in range(10):
    z = np.dot(layer_output, weights[i]) + biases[i]
    layer_output = relu(z)
    print(f"Layer {i+1} output mean:", layer_output.mean())
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
layer_output = x

for i in range(10):
    z = np.dot(layer_output, weights[i]) + biases[i]
    layer_output = sigmoid(z)
    print(f"Layer {i+1} output mean:", layer_output.mean())
```

```python
# Large weight initialization
weights_exploding = []

for i in range(10):
    weights_exploding.append(np.random.randn(20, 20) * 10)

layer_output = x

for i in range(10):
    z = np.dot(layer_output, weights_exploding[i])
    layer_output = relu(z)
    print(f"Layer {i+1} output mean:", layer_output.mean())

```

```python
# Same weights for all neurons
weights_same = []
biases_same = []

for i in range(10):
    weights_same.append(np.ones((20, 20)) * 0.5)
    biases_same.append(np.zeros((1, 20)))

layer_output = x

for i in range(10):
    z = np.dot(layer_output, weights_same[i]) + biases_same[i]
    layer_output = relu(z)
    print(f"Layer {i+1} output (first 3 values):", layer_output[0][:3])
```

```python
def xavier_init(n_in, n_out):
    return np.random.randn(n_in, n_out) * np.sqrt(2 / (n_in + n_out))
```

```python
def he_init(n_in, n_out):
    return np.random.randn(n_in, n_out) * np.sqrt(2 / n_in)
```

```python
weights_he = []

for i in range(10):
    weights_he.append(he_init(20, 20))

layer_output = x

for i in range(10):
    z = np.dot(layer_output, weights_he[i])
    layer_output = relu(z)
    print(f"Layer {i+1} output mean:", layer_output.mean())
```
