```python
import numpy as np

# One input sample with 5 features
x = np.ones((1, 5))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

h = x

print("Vanishing Gradient Demo")
for i in range(10):  # 10 layers
    W = np.random.randn(5, 5) * 0.01   # very small weights
    b = np.zeros((1, 5))
    
    h = sigmoid(h @ W + b)
    print(f"Layer {i+1} output:", h)
```

```python
h = x

print("\nExploding Gradient Demo")
for i in range(10):
    W = np.random.randn(5, 5) * 5   # very large weights
    b = np.zeros((1, 5))
    
    h = h @ W + b
    print(f"Layer {i+1} output:", h)
```

```python
h = x

print("\nSymmetry Problem Demo")
for i in range(10):
    W = np.zeros((5, 5))   # same weights
    b = np.zeros((1, 5))
    
    h = h @ W + b
    print(f"Layer {i+1} output:", h)
```

```python
h = x
n = 5  # number of input neurons

print("\nXavier Initialization Demo")
for i in range(10):
    W = np.random.randn(n, n) * np.sqrt(1 / n)  # Xavier init
    b = np.zeros((1, n))
    
    h = np.tanh(h @ W + b)
    print(f"Layer {i+1} output:", h)
```

```python
def relu(z):
    return np.maximum(0, z)

h = x
n = 5

print("\nHe Initialization Demo")
for i in range(10):
    W = np.random.randn(n, n) * np.sqrt(2 / n)  # He init
    b = np.zeros((1, n))
    
    h = relu(h @ W + b)
    print(f"Layer {i+1} output:", h)
```
