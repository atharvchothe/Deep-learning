```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Flatten
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import to_categorical
```

```python
# Load dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Use only 2000 samples
x_train = x_train[:2000]
y_train = y_train[:2000]

x_test = x_test[:1000]
y_test = y_test[:1000]

# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
baseline_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

baseline_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

baseline_history = baseline_model.fit(
    x_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
plt.plot(baseline_history.history['loss'], label='Train Loss')
plt.plot(baseline_history.history['val_loss'], label='Val Loss')
plt.title("Baseline Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
dropout_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

dropout_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

dropout_history = dropout_model.fit(
    x_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
l2_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(64, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(10, activation='softmax')
])

l2_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

l2_history = l2_model.fit(
    x_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
bn_model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    Dense(64),
    BatchNormalization(),
    tf.keras.layers.Activation('relu'),

    Dense(10, activation='softmax')
])

bn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

bn_history = bn_model.fit(
    x_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
baseline_acc = baseline_model.evaluate(x_test, y_test, verbose=0)[1]
dropout_acc = dropout_model.evaluate(x_test, y_test, verbose=0)[1]
l2_acc = l2_model.evaluate(x_test, y_test, verbose=0)[1]
bn_acc = bn_model.evaluate(x_test, y_test, verbose=0)[1]

print("Baseline Accuracy :", baseline_acc)
print("Dropout Accuracy  :", dropout_acc)
print("L2 Accuracy       :", l2_acc)
print("BatchNorm Accuracy:", bn_acc)
```

```python
plt.figure(figsize=(6,4))
plt.plot(baseline_history.history['loss'], label='Training Loss')
plt.plot(baseline_history.history['val_loss'], label='Validation Loss')
plt.title("Before Regularization: Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(baseline_history.history['accuracy'], label='Training Accuracy')
plt.plot(baseline_history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Before Regularization: Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(bn_history.history['accuracy'], label='Training Accuracy')
plt.plot(bn_history.history['val_accuracy'], label='Validation Accuracy')
plt.title("After Regularization (BatchNorm): Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.plot(baseline_history.history['val_loss'], label='Baseline')
plt.plot(dropout_history.history['val_loss'], label='Dropout')
plt.plot(l2_history.history['val_loss'], label='L2 Regularization')
plt.plot(bn_history.history['val_loss'], label='Batch Normalization')

plt.title("Validation Loss Comparison (All Methods)")
plt.xlabel("Epochs")
plt.ylabel("Validation Loss")
plt.legend()
plt.grid()
plt.show()
```
